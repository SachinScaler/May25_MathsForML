```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

f = lambda x: x**2 # quadartic function: parabola

# h(x, lambda)
## Plot f(x)
x = np.arange(-10, 10, 0.01)
y = f(x)

plt.figure()
plt.plot(x, y)
plt.show()
```

- value of minima F(x) --> 0
- x = 0

### Let's try and see if we can tune this with GD algorithm
- one parameter: once partial derivative

### Helper Functions:

```python
# Function to calculate derivative of f at point x
# 1) Manually calculate derivative
def get_gradient_manual(x):
    return 2*x

#2) first principle: ab intio method
def derivative(x, f, h = 0.00001):
    return (f(x + h) - f(x)) / h

derivative(2, f)
```

```python
get_gradient_manual(2)
```

```python
# Function to perform one update to x
# using gradient and learning rate

def update_step_gd(x, f, lr):
    grad = derivative(x, f)
    new_x = x - lr * grad
    return new_x, grad
```

## Gradient Descent in Action

```python
# Step1: randomly parameter intitialize
x_val = 5
learning_rate = 0.1 # neta
```

```python
# Plot function
plt.figure()

## Plot f(x)
x = np.linspace(-10, 10, 400)
plt.plot(x, f(x))

# Plot initial guess with tangent line
t_x = np.linspace(x_val - 2, x_val + 2, 10)

# -5 and 5**2: point
plt.plot(x_val, f(x_val), "ro", markersize=5, label="Initial Point")  # Initial point
```

```python
# Step2 and 3: Calcuate Gradient
# Update the variable towards opposite of gradient
new_x, grad = update_step_gd(x_val, f, learning_rate)
new_x
```

```python
# Plot initial Tangent line
plt.plot(t_x, grad * t_x + (f(x_val) - grad * x_val), color="black")

# Plot updated point
plt.plot(new_x, f(new_x), "o", markersize=5, color="orange", label="Updated Point")

print("update p", new_x)
print("Grad",grad)
x_val = new_x
plt.legend()
plt.show()
```

```python
# Step 4: Repeat
epochs = 20
```

```python
# Step 1: Initialize parameters
x_val = 5

# hyper parameters
learning_rate = 0.1
num_epochs = 20

# Lists to store x values and function values
x_vals = [x_val]
fx_vals = [f(x_val)]

# Step 2: Gradient Descent Loop
for _ in range(num_epochs):
    new_x, grad = update_step_gd(x_val, f, learning_rate)
    x_vals.append(new_x)
    fx_vals.append(f(new_x))
    x_val = new_x  # Update x_val for the next iteration
```

```python
# Plot function
plt.figure()
x = np.linspace(-10, 10, 400)
plt.plot(x, f(x))

# Plot all points from gradient descent
plt.plot(x_vals, fx_vals, "go-", markersize=4, label="GD Steps")  # Green dots with lines

# Highlight initial and final points
plt.plot(x_vals[0], fx_vals[0], "ro", markersize=6, label="Initial Point")   # Red: Initial
plt.plot(x_vals[-1], fx_vals[-1], "o", color="orange", markersize=6, label="Final Point")  # Orange: Final

# Plot final tangent line
t_x = np.linspace(x_vals[-1] - 2, x_vals[-1] + 2, 10)
plt.plot(t_x, grad * t_x + (f(x_vals[-1]) - grad * x_vals[-1]), color="black", label="Final Tangent")

print("Final x:", x_vals[-1])
print("Final Grad:", grad)

plt.legend()
plt.show()
```
